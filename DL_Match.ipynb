```python
#AND

def AND(a,b):
  input = np.array([a,b])
  weights = np.array([0.4,0.4])
  bias = -0.6
  value = np.sum(input * weights)+ bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
# OR

def OR(a,b):
  input = np.array([a,b])
  weights = np.array([0.4,0.4])
  bias = -0.3
  value = np.sum(input*weights)+bias

  if value <= 0:
    return 0
  else:
    return 1

```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
#NAND

def NAND(a,b):
  input = np.array([a,b])
  weights = np.array([-0.5, -0.5])
  bias = 0.7
  value = np.sum(input * weights)+bias

  if value <= 0:
    return 0
  else:
    return 1
  
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
  result = np.zeros((len(sequences), dimension))

  print(result.shape)
  for i,sequence in enumerate(sequences):
    result[i,sequence]=1
  return result
```
